```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys


spf = wave.open("data/public_test/songs/37303234365f3231.wav", "r")

# Extract Raw Audio from Wav File
signal = spf.readframes(-1)
signal = np.fromstring(signal, "int16")
fs = spf.getframerate()

# If Stereo
# if spf.getnchannels() == 2:
#     print("Just mono files")
#     sys.exit(0)


Time = np.linspace(0, len(signal)/2 / fs, num=len(signal))

plt.figure(1)
plt.title("Signal Wave...")
plt.plot(Time, signal)
plt.show()
```
